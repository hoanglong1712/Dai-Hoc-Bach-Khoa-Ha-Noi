```python
# reading image
import cv2

image = cv2.imread('free-images.jpg')

h, w = image.shape[:2]
print(h, w)
image
```

```python
# resize the imge
# resize() function takes 2 paramenters

resize = cv2.resize(image, (200, 300))
resize
```

```python
# calculate the ratio
ratio = 300/ w

dim = (300, int(h * ratio))

resize_with_aspect = cv2.resize(image, dim)
new_image = resize_with_aspect
resize_with_aspect

```

```python
# rotate the image
# calculate the center of the image

h, w = new_image.shape[:2]

center = (w//2, h //2)

# generate rotation matrix
matrix = cv2.getRotationMatrix2D(center, -45, 1.0)

# performing the affine tranformation
rotated = cv2.warpAffine(new_image, matrix, (w, h))
#rotated
rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
rgb
```

getRotationMatrix2D() takes 3 arguments \\

*   center: center coordinate of the image
*   angle: angle (in degree) by which the image should be rotated
*   scale: scaling factor

It returns a 2*3 matrix consisting of values from alpha  beta. \\
alpha = scale * cos(angle) \\
beta = scale * sin(angle)

M =

\begin{bmatrix}
  α & β & (1 - α) * center.x - \beta * center.y \\
  -β & \alpha & \beta * center.x + (1 - α) * center.y
\end{bmatrix}

warpAffine() \\
the function transforms the source image using rotation matrix \\

(x, y) = src( $M_{11} * X + M_{12} * Y + M_{13}, M_{21} * X + M_{22} * Y + M_{23}  $ )
