```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('phong-nam-valley-cao-bang1.jpg')
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.axis('off')
```

```python
# bien anh 3 mau rgb thanh mang 2D
pixel_values = rgb_img.reshape((-1, 3))
pixel_values = np.float32(pixel_values)

# define criteria
criteria = ( cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,
    100, 0.2)

k = 5

_, labels, centers = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

centers = np.uint8(centers)

# ánh xạ nhãn vào điểm trung tâm của mỗi cụm
segmented_img = centers[labels.flatten()]
segmented_img = segmented_img.reshape(rgb_img.shape)
plt.imshow(segmented_img)
plt.axis('off')
```
