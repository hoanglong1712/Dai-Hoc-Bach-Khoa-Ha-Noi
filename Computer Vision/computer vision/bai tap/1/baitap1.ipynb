```python
import cv2

def load_imgs(img_paths: []) -> []:
  imgs = []
  for item in img_paths:
    imgs.append(cv2.imread(item))
    pass
  return imgs
```

```python
import matplotlib.pyplot as plt

def show_images(imgs: []):
  fig = plt.figure(figsize=(15, 10))
  columns = 3
  rows = int(len(imgs) / columns)

  counter = 1
  for item in imgs:
    fig.add_subplot(rows, columns, counter)
    plt.imshow(item)
    plt.axis('off')
    counter += 1
    pass
  pass
```

```python
images = load_imgs(image_paths)
show_images(images)
```

```python
# Use Opencv to converted the attached image to color spaces: HSV and Lab
hsv_s = [cv2.cvtColor(item, cv2.COLOR_BGR2HSV) for item in images]
lab_s = [cv2.cvtColor(item, cv2.COLOR_BGR2LAB) for item in images]
rgb_s = [cv2.cvtColor(item, cv2.COLOR_BGR2RGB) for item in images]
```

```python
# Then change the V channel or L channel to its maximum.
def to_max(arr, value, index):
  arr_n = arr.copy()
  arr_n[:, :, index] = value
  return arr_n
hsv_max_v = [to_max(item, 255, 2) for item in hsv_s]
lab_max_l = [to_max(item, 255, 0) for item in lab_s]
```

```python
printed_images = []
printed_images.extend(images)
printed_images.extend(hsv_s)
printed_images.extend(lab_s)
printed_images.extend(hsv_max_v)
printed_images.extend(lab_max_l)
printed_images.extend(rgb_s)
```

```python
show_images(printed_images)
```
