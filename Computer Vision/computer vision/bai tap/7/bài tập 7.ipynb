```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('chessboard.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
plt.axis('off')
```

```python
# dng canny pht hin cnh
edges = cv2.Canny(img, 300, 500, apertureSize=3)
edges
```

```python
# s dng HoughLinesP   pht hin cc ng thng
minLineLength = 100
maxLineGap = 80

lines = cv2.HoughLinesP(edges, rho=1, theta=np.pi/180, threshold=100,
                        lines=np.array([]), minLineLength=minLineLength,
                        maxLineGap=maxLineGap)
```

```python
# v cc ng
tmp = img.copy()
for line in lines:
  x1, y1, x2, y2 = line[0]
  cv2.line(tmp, (x1, y1), (x2,y2), (0, 0, 255), 3, cv2.LINE_AA)
  pass

plt.imshow(tmp)
plt.axis('off')
```
