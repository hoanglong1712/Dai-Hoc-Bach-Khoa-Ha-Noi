```python
import cv2
img = cv2.imread('hand.jpg')
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
plt.axis('off')
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
```

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
plt.axis('off')
```

```python
# equalize histogram
ehist = cv2.equalizeHist(img_gray)

fig = plt.figure(figsize=(15, 10))

fig.add_subplot(1, 2, 1)
plt.imshow(ehist)
plt.title("Equalized")
plt.axis('off')

fig.add_subplot(1, 2, 2)
plt.imshow(img_gray)
plt.title("Gray")
plt.axis('off')
```

# How to do it without changing the color image your skin?

* Convert to YUV Color Space:  This step separates the brightness (Y channel) from the color information (U and V channels).
* Equalize the Y Channel:  Y channel (brightness)
* Convert Back to RGB: Finally, convert the YUV image back to RGB

```python
img_yuv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2YUV)
plt.imshow(img_yuv)
plt.axis('off')
```

```python
# Equalize the histogram of the Y channel
img_yuv[:, :, 0] = cv2.equalizeHist(img_yuv[:, :, 0])
img_rgb_equ = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2RGB)
plt.imshow(img_rgb_equ)
plt.axis('off')
```

```python
fig = plt.figure(figsize=(15, 10))

fig.add_subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("RGB")
plt.axis('off')

fig.add_subplot(1, 2, 2)
plt.imshow(img_rgb_equ)
plt.title("EQUALIZED")
plt.axis('off')
```

```python
plt.imsave(fname='gray_equalized.jpg', arr=img_gray)
plt.imsave(fname='rgb_equalized.jpg', arr=img_rgb_equ)
```
