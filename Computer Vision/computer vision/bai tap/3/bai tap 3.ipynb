```python
import cv2
```

```python
def detect(picture, index):
  gray = cv2.cvtColor(picture, cv2.COLOR_BGR2GRAY)
  contours, _ = cv2.findContours(gray, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
  return cv2.drawContours(picture, contours, index, (0, 255,0), 5)
```

```python
picture_1 = cv2.imread('Picture1.png')
picture_2 = cv2.imread('Picture2.png')
```

```python
detect(picture_1.copy(), 2)
```

```python
detect(picture_2.copy(), 1)
```
