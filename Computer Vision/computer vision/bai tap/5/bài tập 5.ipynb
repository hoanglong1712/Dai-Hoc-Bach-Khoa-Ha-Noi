```python
import cv2
import numpy as np
```

```python
# đọc ảnh
cup = cv2.imread('cup.jpg')
background = cv2.imread('background.jpg')

cup_gray = cv2.cvtColor(cup, cv2.COLOR_BGR2GRAY)
background_gray = cv2.cvtColor(background, cv2.COLOR_BGR2GRAY)
```

```python
import matplotlib.pyplot as plt
plt.imshow(cup_gray)
plt.axis('off')
```

```python
import matplotlib.pyplot as plt
plt.imshow(background_gray)
plt.axis('off')
```

```python
result = cv2.subtract(cup_gray, background_gray)
import matplotlib.pyplot as plt
plt.imshow(result)
plt.axis('off')
```

```python
result = cv2.subtract(background_gray, cup_gray)
import matplotlib.pyplot as plt
plt.imshow(result)
plt.axis('off')
```

```python
# định nghĩa 1 cái nhân hay còn gọi là 1 ma  trận
kernel = np.ones((5, 5), np.uint8)
kernel
```

```python
# opening
opening = cv2.morphologyEx(background_gray, cv2.MORPH_OPEN, kernel)
import matplotlib.pyplot as plt
plt.imshow(opening)
plt.axis('off')
```

```python
result = cv2.subtract(cup_gray, opening)
import matplotlib.pyplot as plt
plt.imshow(result)
plt.axis('off')
```

```python
# closing
closing = cv2.morphologyEx(background_gray, cv2.MORPH_CLOSE, kernel)
import matplotlib.pyplot as plt
plt.imshow(closing)
plt.axis('off')
```

```python
result = cv2.subtract(cup_gray, closing)
import matplotlib.pyplot as plt
plt.imshow(result)
plt.axis('off')
```
